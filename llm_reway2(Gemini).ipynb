```python
import fitz  # PyMuPDF
from google.colab import files

# Input the path to your PDF file
pdf_path = "/content/EWaste_Sep11_892011.pdf"

# Open the PDF file
pdf_document = fitz.open(pdf_path)

# Iterate through each page and extract text
text = ""
for page_number in range(pdf_document.page_count):
    page = pdf_document[page_number]
    text += page.get_text()

# Print the concatenated text
print(text)



```

```python
context = "\n".join(page.get_text("text") for page in pdf_document)
```

```python
print("The total number of words in the context:", len(context))
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=5000, chunk_overlap=200)
context = "\n\n".join(page.get_text("text") for page in pdf_document)
```

```python
texts = text_splitter.split_text(context)
```

```python
print(len(texts))
```

```python
texts[0]
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
!export GOOGLE_API_KEY=AIzaSyDBN--zTGlBpmY-agZBTsqGJ0Po6yFwvFE
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model = "models/embedding-001", google_api_key = "AIzaSyDBN--zTGlBpmY-agZBTsqGJ0Po6yFwvFE")
```

```python
vector_index = Chroma.from_texts(texts, embeddings).as_retriever()
```

```python
question = "what is E-waste"
docs = vector_index.get_relevant_documents(question)
```

```python
docs
```

```python
prompt_template = """
  Answer the question as detailed as possible from the provided context, make sure to provide all the details, if the answer is not in
  provided context just say, "answer is not available in the context", don't provide the wrong answer\n\n
  Context:\n {context}?\n
  Question: \n{question}\n

  Answer:
"""

prompt = PromptTemplate(template = prompt_template, input_variables = ["context", "question"])
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-pro",google_api_key="AIzaSyDBN--zTGlBpmY-agZBTsqGJ0Po6yFwvFE",
                             temperature=0.3)
```

```python
chain = load_qa_chain(model, chain_type="stuff", prompt=prompt)
```

```python
response = chain(
    {"input_documents":docs, "question": question}
    , return_only_outputs=True)
```
